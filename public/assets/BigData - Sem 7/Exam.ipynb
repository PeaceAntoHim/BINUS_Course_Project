```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data simulasi produksi minyak (dalam ribuan barel per hari) berdasarkan regional dan bulan
data = {
    'Regional': ['Sumatera', 'Kalimantan', 'Papua', 'Jawa', 'Sulawesi'],
    'Januari': [38.894, 15.466, 18.643, 5.467, 39.329],
    'Februari': [68.233, 47.828, 39.693, 30.774, 69.858],
    'Maret': [10.615, 47.828, 79.574, 32.598, 45.258],
}

df = pd.DataFrame(data)
df_long = pd.melt(df, id_vars=['Regional'], var_name='Bulan', value_name='Produksi')

# Visualisasi menggunakan seaborn barplot
plt.figure(figsize=(12, 6))
sns.barplot(data=df_long, x='Regional', y='Produksi', hue='Bulan', palette='Set2')

plt.title('Produksi Minyak per Hari per Regional (Januari - Maret 2024)', fontsize=14)
plt.ylabel('Produksi (ribu barel/hari)')
plt.xlabel('Regional')
plt.legend(title='Bulan')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
