```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Q4 Data - Company Performance
quarters = pd.date_range(start='2020-01-01', periods=20, freq='QE')
revenue = np.linspace(500, 900, 20) + np.random.normal(0, 20, 20)
profit = revenue * (0.1 + 0.02 * np.sin(np.linspace(0, 6.28, 20))) + np.random.normal(0, 5, 20)

plt.figure()
plt.plot(quarters, revenue, label='Revenue (in millions)')
plt.plot(quarters, profit, label='Profit (in millions)')
plt.title('Company Performance Over Time')
plt.xlabel('Quarter')
plt.ylabel('Millions')
plt.legend()
plt.tight_layout()
plt.show()

# Q4 Data - Credit Scoring
scores = np.random.normal(650, 50, 1000)
plt.figure()
plt.hist(scores, bins=20)
plt.title('Distribution of Credit Scores')
plt.xlabel('Credit Score')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()

# Q5 Data - Comparison
regions = ['North', 'South', 'East', 'West']
sales = [200, 150, 180, 220]
plt.figure()
plt.bar(regions, sales)
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales (Units)')
plt.tight_layout()
plt.show()

# Q5 Data - Composition
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sizes = [30, 25, 20, 25]
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Market Share by Product')
plt.tight_layout()
plt.show()

# Q5 Data - Distribution
ages = np.random.normal(35, 10, 500)
plt.figure()
plt.hist(ages, bins=15)
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Number')
plt.tight_layout()
plt.show()

# Q5 Data - Relationship
spend = np.random.uniform(50, 200, 50)
rev = 2 * spend + np.random.normal(0, 10, 50)
plt.figure()
plt.scatter(spend, rev)
plt.title('Marketing Spend vs Revenue')
plt.xlabel('Marketing Spend (Thousands)')
plt.ylabel('Revenue (Thousands)')
plt.tight_layout()
plt.show()
```
