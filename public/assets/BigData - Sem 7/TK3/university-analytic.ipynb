```python
import pandas as pd
import matplotlib.pyplot as plt
import ace_tools_open as tools  # Use ace_tools for display_dataframe_to_user

# Sample synthetic data for analysis
years = [2020, 2021, 2022, 2023, 2024]
enrolled = [1200, 1300, 1250, 1400, 1500]
applicants = [3000, 3200, 3100, 3500, 3800]
acceptance_rate = [enrolled[i] / applicants[i] * 100 for i in range(len(years))]

gpa_avg = [3.2, 3.25, 3.3, 3.35, 3.4]
graduation_rate = [75, 78, 80, 82, 85]

months = pd.date_range(start='2024-01', periods=12, freq='M').strftime('%b')
library_visits = [500, 450, 480, 520, 610, 630, 600, 580, 620, 640, 660, 690]
lab_utilization = [65, 60, 62, 70, 75, 80, 78, 76, 79, 82, 85, 88]

# Create DataFrame
df_summary = pd.DataFrame({
    'Year': years,
    'Applicants': applicants,
    'Enrolled': enrolled,
    'Acceptance Rate (%)': [round(x, 1) for x in acceptance_rate],
    'Average GPA': gpa_avg,
    'Graduation Rate (%)': graduation_rate
})

# Display DataFrame using ace_tools
tools.display_dataframe_to_user(name="Summary Metrics by Year", dataframe=df_summary)

# Plot figures
plt.figure()
plt.plot(years, enrolled, marker='o')
plt.title('Enrollment Trend (2020-2024)')
plt.xlabel('Year')
plt.ylabel('Number of Enrolled Students')
plt.tight_layout()
plt.show()

plt.figure()
plt.plot(years, acceptance_rate, marker='s')
plt.title('Acceptance Rate Trend (2020-2024)')
plt.xlabel('Year')
plt.ylabel('Acceptance Rate (%)')
plt.tight_layout()
plt.show()

plt.figure()
plt.plot(years, gpa_avg, marker='^')
plt.title('Average GPA Trend (2020-2024)')
plt.xlabel('Year')
plt.ylabel('GPA')
plt.tight_layout()
plt.show()

plt.figure()
plt.plot(years, graduation_rate, marker='v')
plt.title('Graduation Rate Trend (2020-2024)')
plt.xlabel('Year')
plt.ylabel('Graduation Rate (%)')
plt.tight_layout()
plt.show()

# Facilities usage plots
plt.figure()
plt.bar(months, library_visits)
plt.title('Library Visits per Month (2024)')
plt.xlabel('Month')
plt.ylabel('Number of Visits')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure()
plt.plot(months, lab_utilization, marker='.')
plt.title('Lab Utilization Rate per Month (2024)')
plt.xlabel('Month')
plt.ylabel('Utilization Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
